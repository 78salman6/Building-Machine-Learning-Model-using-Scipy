```python
import pandas as pd
```

```python
auto_data = pd.read_csv('../data/imports-85.data', sep=r'\s*,\s*', engine='python')
auto_data
```

```python
import numpy as np
auto_data = auto_data.replace('?', np.nan)
auto_data.head()
```

```python
auto_data.describe()
```

```python
auto_data['price'].describe()
```

```python
auto_data['price'] = pd.to_numeric(auto_data['price'], errors = 'coerce')
```

```python
auto_data = auto_data.drop('normalized-losses', axis = 1)
auto_data.head()
```

```python
auto_data.describe()
```

```python
auto_data['horsepower'].describe()
```

```python
auto_data['horsepower'] = pd.to_numeric(auto_data['horsepower'], errors = 'coerce')
```

```python
auto_data['horsepower'].describe()
```

```python
auto_data['num-of-cylinders'].describe()
```

```python
cylinders_dict = { 'two' : 2,
                  'three' : 3,
                  'four' : 4,
                  'five' : 5,
                  'six' : 6,
                  'eight' : 8,
                  'twelve' : 12}
auto_data['num-of-cylinders'].replace(cylinders_dict, inplace=True)
auto_data.head()
```

```python
auto_data['num-of-cylinders'].head()
```

```python
auto_data = pd.get_dummies(auto_data, 
                          columns = ['make',
                                     'fuel-type',
                                      'aspiration',
                                     'num-of-doors',
                                     'body-style',
                                     'drive-wheels',
                                      'engine-location',
                                     'engine-type',
                                      'fuel-system'])
auto_data.head()
```

```python
auto_data = auto_data.dropna()
auto_data
```

```python
from sklearn.model_selection import train_test_split

X = auto_data.drop('price', axis=1)

#Taking the labels(Price)
Y = auto_data['price']

#Splitting into 80% for training set and 20% for testing set so we can see our accuracy
X_train, x_test, Y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression

linear_model = LinearRegression()
linear_model.fit(X_train, Y_train)
```

```python
#R2 score
linear_model.score(X_train, Y_train)
```

```python
linear_model.coef_
```

```python
predictors = X_train.columns
coef = pd.Series(linear_model.coef_, predictors).sort_values()
print(coef)
```

```python
y_predict = linear_model.predict(x_test)
```

```python
%pylab inline
pylab.rcParams['figure.figsize'] = (15, 6)

plt.plot(y_predict, label='Predicted')
plt.plot(y_test.values, label='Actual')
plt.ylabel('Price')

plt.legend()
plt.show()
```

```python
r_square = linear_model.score(x_test, y_test)
r_square
```

```python
from sklearn.metrics import mean_squared_error

linear_model_mse = mean_squared_error(y_predict, y_test)
linear_model_mse
```

```python
import math
math.sqrt(linear_model_mse)
```

# Demo Lasso and Ridge Regression using Scikit Learn
## Lasso

```python
from sklearn.linear_model import Lasso
# normalize = True means "Centre the data by subtracting the mean and Dividing by the L2 norm
lasso_model = Lasso(alpha=5, normalize=True)
lasso_model.fit(X_train, Y_train)
```

```python
lasso_model.score(X_train, Y_train)
```

```python
coef = pd.Series(lasso_model.coef_, predictors).sort_values()
```

```python
# Large negative values means these features are inversely correlated to output price
print(coef)
```

```python
y_predict = lasso_model.predict(x_test)
```

```python
pylab.rcParams['figure.figsize'] = (15, 6)
plt.plot(y_predict, label='Predicted')
plt.plot(y_test.values, label='Actual')

plt.legend()
plt.show()
```

```python
r_square = lasso_model.score(x_test, y_test)
r_square
```

```python
# Mean Squared error is much better than Simple Regression for which it was around (5800)
lasso_model_mse = mean_squared_error(y_predict, y_test)
math.sqrt(lasso_model_mse)
```
